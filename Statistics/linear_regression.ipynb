Линейная регрессия - это метод определения взаимосвязи между несколькими количественными переменными.
Если взаимосвязь можно описать с помощью одной прямой, то к ней можно применить линейную регрессию.
Формула прямой описывается как: y = b_0 + b_1 * x,
где b_1 = (sd_y)/(sd_x) * r_(x, y), b_0 = y_mean - b_1 * x_mean;

```python
import numpy as np
import matplotlib.pyplot as plt

# Значения случайных величин
x = np.array([1, 3, 2, 3, 4, 5, 6, 7, 10, 8])
y = np.array([10, 11, 10, 9, 13, 15, 7, 18, 22, 16]);

# Считаем коэффициенты
def cov(x, y):
    assert x.size == y.size
    return ((x - x.mean()) * (y - y.mean())).sum() / (x.size - 1)

def cor(x, y):
    return cov(x, y) / (np.std(x, ddof=1) * np.std(y, ddof=1))

b1 = (y.std() / x.std()) * cor(x, y)
b0 = y.mean() - b1 * x.mean()

# Регрессионная прямая
f = lambda el: (b0 + b1 * el)
y1 = np.array([f(el) for el in x])

# Рисуем график
plt.plot(x, y, "ro")
plt.plot(x, y1, "g")
```
