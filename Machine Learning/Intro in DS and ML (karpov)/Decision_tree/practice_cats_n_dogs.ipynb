```python
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn import tree
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df_cats_n_dogs = pd.read_csv("../../../Datasets/dogs_n_cats.csv")
df_cats_n_dogs.head()
```

```python
X = df_cats_n_dogs.drop(["Вид"], axis=1)
y = df_cats_n_dogs["Вид"]

print(X.head(), y.head(), sep="\n")
```

### Определим самые лучшие параметры для дерева

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=0)

scores = pd.DataFrame()
max_depth_range = range(1, 100)

for max_depth in max_depth_range:
    clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth=max_depth)
    clf.fit(X_train, y_train)
    train_score = clf.score(X_train, y_train)
    test_score  = clf.score(X_test, y_test)

    mean_cross_score = cross_val_score(clf, X_train, y_train, cv=5).mean()

    temp = pd.DataFrame({
        "depth": [max_depth],
        "train_score": [train_score],
        "test_score": [test_score],
        "mean_cross_score": [mean_cross_score]
    }) 

    scores = scores.append(temp)

scores.head(20)
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy", max_depth=5)
clf.fit(X, y)
```
