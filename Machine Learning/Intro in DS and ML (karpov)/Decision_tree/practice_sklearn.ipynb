```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
heart_disease_data = pd.read_csv("../../../Datasets/train_data_tree.csv")
heart_disease_data.head()
```

```python
X = heart_disease_data.drop(["num"], axis=1)
y = heart_disease_data.num
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
# Посчитаем энтропию
left_leaf  = clf.tree_.children_left[0]
right_leaf = clf.tree_.children_right[0]

entropy_root  = clf.tree_.impurity[0]
entropy_left  = clf.tree_.impurity[left_leaf]
entropy_right = clf.tree_.impurity[right_leaf]

samples_root  = clf.tree_.n_node_samples[0]
samples_left  = clf.tree_.n_node_samples[left_leaf]
samples_right = clf.tree_.n_node_samples[right_leaf]

print(f"Information Gain = {entropy_root - samples_left/samples_root*entropy_left - samples_right/samples_root*entropy_right}")
```
